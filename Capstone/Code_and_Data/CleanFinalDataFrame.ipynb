```python
import pandas as pd
```

```python
slide = pd.read_csv('DataParts/SLIDE_FULL.csv')
noslide = pd.read_csv('DataParts/NOSLIDE_FULL.csv')
```

```python
slide.describe()
```

```python
noslide.describe()
```

```python
frames = [slide,noslide]
landslides = pd.concat(frames,ignore_index=True)
```

```python
landslides.fillna(0, inplace = True)
```

```python
landslides["Slope"] = landslides["Northeast"] + landslides["North"] + landslides['Center'] + landslides['West']  + landslides['Northwest'] + landslides['SouthEast'] + landslides['East'] + landslides['South'] + landslides['SouthWest']
list(landslides)
```

```python
landslides['DATE'] = pd.to_datetime(landslides['DATE'])
```

```python
landslides['COVERAGE_1'] = landslides['nlcd_2011_'].where(pd.to_datetime(landslides['DATE']).dt.year >= 2011)
landslides['COVERAGE_2'] = landslides['nlcd_2011_'].where((pd.to_datetime(landslides['DATE']).dt.year >= 2006) & (pd.to_datetime(landslides['DATE']).dt.year < 2011))
landslides['COVERAGE_3'] = landslides['nlcd_2011_'].where(pd.to_datetime(landslides['DATE']).dt.year < 2006)
```

```python
landslides.fillna(0, inplace = True)
```

```python
landslides['COVERAGE'] = landslides['COVERAGE_1'] + landslides['COVERAGE_2'] + landslides['COVERAGE_3']
```
