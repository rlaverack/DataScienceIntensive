```python
import pandas as pd
```

```python
landslides = pd.read_csv('landslides_lat_long.csv')
landslides.head()
```

```python
geo_info = pd.read_csv('GeologicDataLandlsides.csv')
```

```python
landslides_geo = landslides.merge(geo_info,on=u'UNIQUE_ID')
```

```python
landslides_geo.head()
```

```python
slope = pd.read_csv('SlopeData.csv')
slope.head()
```

```python
landslides_sl = landslides_geo.merge(slope,on=u'UNIQUE_ID')
```

```python
landslides_sl.head()
```

```python
landslides_sl.to_csv('landslides_noweather.csv')
```
